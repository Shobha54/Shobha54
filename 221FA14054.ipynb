```python
# prompt: from the above data in the file give me linear regression between TS1	TS2	in python and also represent the title in that graph

import matplotlib.pyplot as plt
import numpy as np

# Assuming 'TS1' and 'TS2' are columns in your DataFrame 'df'
x = df['TS1']
y = df['TS2']

# Calculate the linear regression line
coefficients = np.polyfit(x, y, 1)
polynomial = np.poly1d(coefficients)
regression_line = polynomial(x)

# Plot the data points and the regression line
plt.scatter(x, y, label='Data Points')
plt.plot(x, regression_line, color='red', label='Linear Regression')

# Add title and labels
plt.title('Linear Regression of TS1 vs TS2')
plt.xlabel('TS1')
plt.ylabel('TS2')
plt.legend()

# Show the plot
plt.show()
```
